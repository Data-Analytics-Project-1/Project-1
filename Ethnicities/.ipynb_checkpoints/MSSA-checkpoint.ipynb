```python
import pandas as pd
import requests
import gmaps

from config import g_key 
```

```python
serviceAreas_df = pd.DataFrame(pd.read_csv("MSSA_Detail.csv"))
serviceAreas_df.head()  
```

```python
serviceAreas_df.drop(columns=['OBJECTID', 'STATE_FIPS', 'MSSA_ID', 'TRACT_ID', 'TRACT_CODE', 'TRACT', 'CENSUS_KEY', 'COUNTY',
                             'MSSA_NAME', 'Shape_STArea_1', 'Shape_STLength_1', 'Shape__Area', 'Shape__Length', 'AGE_UNDER5', 'PCT_UNDER5', 'AGE_UNDR18',
                             'PCT_UNDR18', 'AGE_18_64', 'PCT_18_64', 'AGE_65OVER', 'PCT_65OVER', 'ELD_RATIO', 'YTH_RATIO', 'PCT_NHS_BLACK',
                              'TOTALPOVPOP', 'POP100POV', 'PCT100POV', 'POP200POV', 'PCT200POV', 'PCT_WHITE', 'NHS_WHITE', 'PCT_NHS_WHITE', 
                              'PCT_BLACK', 'NHS_BLACK', 'PCT_ASIAN', 'NHS_ASIAN', 'PCT_NHS_ASIAN', 'PCT_AMIND_ESK', 'NHS_AMIND_ESK', 'PCT_NHS_AMIND_ESK',
                              'PCT_ISLAND', 'NHS_ISLANDER', 'PCT_NHS_ISLANDER', 'PCT_OTHER', 'NHS_OTHER_RACE', 'PCT_NHS_OTHER_RACE', 'PCT_MULTI', 
                              'NHS_MULTI_RACE', 'PCT_NHS_MULTI_RACE', 'PCT_HSPNC', 'POP_CIV', 'AREA_SQMI'], inplace=True)

serviceAreas_df.dropna(inplace=True)

serviceAreas_df.head()
```

```python
serviceAreas_df = serviceAreas_df.groupby('CNTY_FIPS').agg({'POP':'sum', 'WHITE':'sum', 'BLACK':'sum',
                                                           'ASIAN':'sum', 'AMIND_ESK':'sum', 'ISLANDER':'sum', 'OTHER_RACE':'sum', 'MULTI_RACE':'sum', 
                                                            'HISPANIC':'sum'})

```

```python
serviceAreas_df.head(15)
```

```python
serviceAreas_df.columns
```

```python
serviceAreas_df['MINORITY'] = serviceAreas_df['BLACK'] + serviceAreas_df['ASIAN'] + serviceAreas_df['AMIND_ESK'] + serviceAreas_df['ISLANDER'] + serviceAreas_df['OTHER_RACE'] + serviceAreas_df['MULTI_RACE'] + serviceAreas_df['HISPANIC']

serviceAreas_df.head(10)
```

```python
serviceAreas_df.drop(columns=['BLACK', 'ASIAN', 'AMIND_ESK',
       'ISLANDER', 'OTHER_RACE', 'MULTI_RACE', 'HISPANIC'], inplace=True)

serviceAreas_df.head(10)
```

```python
serviceAreas_df['WHITE %'] = (serviceAreas_df['WHITE']/serviceAreas_df['POP'])*100
serviceAreas_df['MINORITY %'] = (serviceAreas_df['MINORITY']/serviceAreas_df['POP'])*100
```

```python
m = serviceAreas_df['MINORITY %'].max()
m
```

```python
serviceAreas_df.head(10)
```

```python
latlng_df = pd.DataFrame(pd.read_csv("CenPop2010CountyCenters.csv"))
latlng_df.head()
```

```python
gmaps.configure(api_key=g_key)

fig = gmaps.figure()
```

```python
locations = latlng_df[["LATITUDE", "LONGITUDE"]].astype(float)



heat_layer_white = gmaps.heatmap_layer(locations, weights=serviceAreas_df['WHITE %'], 
                                 dissipating=False, gradient=['rgba(0, 252, 255, 0.05)', 'rgba(0, 242, 255, 0.59)', 'rgba(0, 210, 255, 1)'], 
                                 max_intensity=50, point_radius = 1)

fig.add_layer(heat_layer_white)


heat_layer_minority = gmaps.heatmap_layer(locations, weights=serviceAreas_df['MINORITY %'], 
                                 dissipating=True, gradient=['rgba(255, 130, 0, 0.38)', 'rgba(255, 130, 0, 0.65)', 'rgba(255, 130, 0, 1)'], max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer_minority)



fig
```
